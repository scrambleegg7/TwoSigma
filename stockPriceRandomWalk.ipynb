```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random
     
def Step(n_dim):
    move = [0 for i in xrange(n_dim)]
    dimension_to_move = random.randint(0, n_dim-1)
    direction = 1 if random.random() > 0.5 else -1
    move[dimension_to_move] = direction
    return move
     
def RandomWalkMulti(n_steps, n_dim):
    return [Step(n_dim) for i in xrange(n_steps)]
```

```python
rw = RandomWalkMulti(100,1)
len(rw)
```

```python
rw_cumsum = np.cumsum(rw)
```

```python
plt.plot(range(len(rw)),rw_cumsum,'x')
```

```python
from scipy.stats import norm
 
#Define Variables
T = 10000 #Number of trading days (we also run at 1000 for the purposes of comparison).
mu = 0.12 #Return
vol = 0.1 #Volatility
 
daily_mu = mu / np.float(T)
daily_vol = vol / np.sqrt(T) 

print("daily rate %.7f" % daily_mu)
print("daily vol %7f" % daily_vol)
    
daily_returns=np.random.normal(daily_mu,daily_vol,T)

print(np.mean(daily_returns))
print(np.std(daily_returns))
 
price_list = [200]
for x in daily_returns:
    price_list.append(price_list[-1]*  (1. + x) )
 
#Generate Plots
plt.plot(range(len(price_list)) , price_list,'x')
#plt.hist(daily_returns-1, 100) #Note that we run the line plot and histogram separately, not simultaneously
```

```python
pr_list = np.array(price_list)
```

```python
y = np.log((pr_list[1:] / pr_list[:-1]))
```

```python
plt.plot(range(len(y)), np.cumsum(y),'x',c='r')
y_cumsum = np.cumsum(y)
```
